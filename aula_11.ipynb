pratique 100

```python
import matplotlib.pyplot as plt

gdp_cap = [500, 1000, 5000, 20000, 50000]
life_exp = [45, 60, 70, 75, 80]

plt.scatter(gdp_cap, life_exp)
plt.xscale('log')

xlab = 'PIB per capita'
ylab = 'Expectativa de vida'
title = 'Expectativa de vida vs PIB per capita'

plt.xlabel(xlab)
plt.ylabel(ylab)
plt.title(title)
plt.show()
```

pratique 101

```python
import matplotlib.pyplot as plt

tick_val = [1000, 10000, 100000]
tick_lab = ['1k', '10k', '100k']

plt.scatter(gdp_cap, life_exp)
plt.xscale('log')

xlab = 'PIB per capita'
ylab = 'Expectativa de vida'
title = 'Expectativa de vida vs PIB per capita'

plt.xticks(tick_val, tick_lab)
plt.xlabel(xlab)
plt.ylabel(ylab)
plt.title(title)
plt.show()
```
