pratique 86

```python
import matplotlib.pyplot as plt

ano = [2020, 2021, 2022, 2023, 2024]
peso = [98, 95, 92, 88, 85]


print("Último ano:", ano[-1])
print("Último peso:", peso[-1])


plt.plot(ano, peso, marker='o', color='blue')


plt.xlabel("Ano")
plt.ylabel("Peso (kg)")
plt.title("Evolução do peso ao longo dos anos")

plt.show()
```

pratique 87

```python
import matplotlib.pyplot as plt
gdp_cap = [500, 1000, 5000, 20000, 50000]
life_exp = [45, 60, 70, 75, 80]

plt.scatter(gdp_cap, life_exp)

plt.xscale('log')
plt.xlabel('PIB per capita')
plt.ylabel('Expectativa de vida')
plt.title('Expectativa de vida vs PIB per capita')

plt.show()
```
